QTextEdit 是 Qt 中的一个小部件（Widget），用于创建多行文本编辑框。它允许用户输入、编辑和显示多行文本内容，例如文本文件、日志、富文本文档等。以下是 QTextEdit 的一些常见用法和属性：

1. **创建 QTextEdit**：
   要创建一个 QTextEdit，您可以使用 `QTextEdit` 类的构造函数。例如：

   ```python
   text_edit = QTextEdit()
   ```

2. **设置和获取文本**：
   您可以使用 `setText()` 方法来设置 QTextEdit 的文本内容，以及使用 `toPlainText()` 方法来获取当前文本内容。

   ```python
   text_edit.setText("这是一个多行文本编辑框")
   current_text = text_edit.toPlainText()
   ```

3. **富文本编辑**：
   QTextEdit 支持富文本编辑，允许用户设置文本的样式（例如粗体、斜体、颜色等）。您可以使用 `setHtml()` 方法来设置富文本文档。

   ```python
   text_edit.setHtml("<h1>这是一个标题</h1><p>这是一个段落。</p>")
   ```

4. **信号与槽**：
   QTextEdit 可以发出多种信号，例如 `textChanged` 信号，用于在文本发生变化时触发相应的操作。您可以连接 QTextEdit 的信号到一个槽函数，以便在文本发生变化时执行特定的操作。

   ```python
   text_edit.textChanged.connect(self.text_changed)
   
   def text_changed(self):
       print("文本发生变化")
   ```

5. **设置输入限制**：
   您可以使用 `setMaxLength()` 方法来设置输入的最大长度限制。此外，还可以使用 `setAcceptRichText()` 方法来允许或禁止富文本输入。

   ```python
   text_edit.setMaxLength(100)  # 设置最大长度为100个字符
   text_edit.setAcceptRichText(False)  # 禁止富文本输入
   ```

6. **滚动条**：
   如果 QTextEdit 中的文本内容超出了可见区域，它会自动添加滚动条以允许用户滚动查看全部内容。

这只是 QTextEdit 的一些常见用法。它是用户界面中常见的元素，用于多行文本输入和编辑。您可以根据您的应用程序需求对 QTextEdit 进行更多的自定义和样式设置。

# 示例

以下是一个简单的示例，演示如何创建一个 QTextEdit 用于多行文本输入和显示。在这个示例中，我们还创建了一个 QLabel，用于显示 QTextEdit 中的字符数。

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QTextEdit, QVBoxLayout, QLabel

class MyWindow(QWidget):
    def __init__(self):
        super().__init__()

        # 设置窗口标题
        self.setWindowTitle("QTextEdit 示例")

        # 创建垂直布局管理器
        layout = QVBoxLayout(self)

        # 创建 QTextEdit 用于多行文本输入
        self.text_edit = QTextEdit(self)
        layout.addWidget(self.text_edit)

        # 创建 QLabel 用于显示字符数
        self.label = QLabel(self)
        layout.addWidget(self.label)

        # 连接 QTextEdit 的信号与槽
        self.text_edit.textChanged.connect(self.update_character_count)

        # 将布局设置为窗口的主布局
        self.setLayout(layout)

    def update_character_count(self):
        # 获取 QTextEdit 中的文本内容并显示字符数
        text = self.text_edit.toPlainText()
        character_count = len(text)
        self.label.setText(f"字符数：{character_count}")

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MyWindow()
    window.show()
    sys.exit(app.exec_())
```

在这个示例中，我们创建了一个 QTextEdit 用于多行文本输入，用户可以在其中输入文本。还创建了一个 QLabel，用于显示 QTextEdit 中的字符数。当用户输入文本时，QTextEdit 会发出 `textChanged` 信号，我们连接这个信号到 `update_character_count` 槽函数，以更新字符数的显示。这个示例演示了如何使用 QTextEdit 创建多行文本输入和显示。
