QLineEdit 是 Qt 中的一个小部件（Widget），用于创建文本输入框。它通常用于接收用户输入的文本信息，例如用户名、密码、搜索关键字等。以下是 QLineEdit 的一些常见用法和属性：

1. **创建 QLineEdit**：
   要创建一个 QLineEdit，您可以使用 `QLineEdit` 类的构造函数。例如：

   ```python
   line_edit = QLineEdit()
   ```

2. **设置和获取文本**：
   您可以使用 `setText()` 方法来设置 QLineEdit 的文本内容，以及使用 `text()` 方法来获取当前文本内容。

   ```python
   line_edit.setText("这是一个文本输入框")
   current_text = line_edit.text()
   ```

3. **设置占位文本**：
   您可以使用 `setPlaceholderText()` 方法来设置占位文本，这是在用户未输入文本时显示的灰色提示文本。

   ```python
   line_edit.setPlaceholderText("请输入文本...")
   ```

4. **信号与槽**：
   QLineEdit 可以发出多种信号，例如 `textChanged` 信号，用于在文本发生变化时触发相应的操作。您可以连接 QLineEdit 的信号到一个槽函数，以便在文本发生变化时执行特定的操作。

   ```python
   line_edit.textChanged.connect(self.text_changed)
   
   def text_changed(self, new_text):
       print("文本发生变化:", new_text)
   ```

5. **设置输入限制**：
   您可以使用 `setMaxLength()` 方法来设置输入的最大长度限制。此外，还可以使用正则表达式来限制输入的内容。

   ```python
   line_edit.setMaxLength(10)  # 设置最大长度为10个字符
   line_edit.setValidator(QRegExpValidator(QRegExp("[0-9]+")))  # 限制只能输入数字
   ```

6. **密码模式**：
   如果您需要创建一个密码输入框，可以将 QLineEdit 设置为密码模式，以便隐藏输入的文本。

   ```python
   line_edit.setEchoMode(QLineEdit.Password)
   ```

这只是 QLineEdit 的一些常见用法。它是用户界面中常见的元素，用于接收和显示文本输入。您可以根据您的应用程序需求对 QLineEdit 进行更多的自定义和样式设置。

# 示例

以下是一个简单的示例，演示如何创建一个 QLineEdit 用于接收用户输入的文本，并在用户输入文本后显示输入的文本。在这个示例中，我们还使用了一个 QPushButton，用户可以点击按钮以清除文本输入框中的文本。

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLineEdit, QPushButton, QVBoxLayout

class MyWindow(QWidget):
    def __init__(self):
        super().__init__()

        # 设置窗口标题
        self.setWindowTitle("QLineEdit 示例")

        # 创建垂直布局管理器
        layout = QVBoxLayout(self)

        # 创建文本输入框
        self.line_edit = QLineEdit(self)
        self.line_edit.setPlaceholderText("请输入文本...")
        layout.addWidget(self.line_edit)

        # 创建按钮以清除文本
        clear_button = QPushButton("清除文本", self)
        clear_button.clicked.connect(self.clear_text)
        layout.addWidget(clear_button)

        # 将布局设置为窗口的主布局
        self.setLayout(layout)

    def clear_text(self):
        # 清除文本输入框中的文本
        self.line_edit.clear()

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MyWindow()
    window.show()
    sys.exit(app.exec_())
```

在这个示例中，我们创建了一个文本输入框（QLineEdit），用户可以在其中输入文本。还创建了一个按钮（QPushButton），用户可以单击以清除文本输入框中的文本。当用户输入文本后，文本将显示在文本输入框中。当用户单击按钮时，文本将被清除。这个示例演示了如何使用 QLineEdit 创建文本输入框，并使用 QPushButton 来执行相应的操作。
